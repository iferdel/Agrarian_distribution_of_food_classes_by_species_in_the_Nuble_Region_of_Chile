# Agrarian distribution of food classes by species in the Ñuble Region of Chile

This notebooks comes from a personal project, therefore **the compromisable data is avoided** (names, personal ids, etc.). All the data being used here has been created artificially by the code uploaded as the notebook xlsx_generator_random.xlsx which creates randomly ids (*rut*), food classes distribution per id and the coordenates of every agrarian person for every species of food class in the region. 

**We want to have a visual representation of the vineyards in the Ñuble Region of Chile that have incentives from the state and acknowledge them by their acres.**

In order to correctly visualize the vineyards for the whole region we would want to classify them by a parameter, in this case we will import and superpose the artificial agrarain data with the real spatial data provided by the *Biblioteca del Congreso Nacional de Chile* (**https://www.bcn.cl/siit/mapas_vectoriales**) which shows the division of every region of the country by their communes. In our case, we will use the communes as the parameter for visualization of the data.

## Raw data manipulation

First of all, the files containing the data will be uploaded and analysed/cleaned in order to define which column will be used to merge them and avoid any conflict in the merge itself related to tiping or having different formats.

```python
#read and clean on columns to merge
df1 = pd.read_excel('data_1.xlsx')
df1.columns = df1.columns.str.strip()
df1.head(3)
```

```python
len(df1)
```

The personal ID (*rut*) will be the column being used to merge all the datafiles so we normalize the typing of every personal ID so we create a function that strip from blank spaces and remove dots if any or add a dash symbol whenever it is not contained, as the pattern we want is XXXXXXXX-X.

```python
def id_cleaner(df, input_colname, output_colname):
    
    df[input_colname] = df.loc[:, input_colname].str.strip()
    
    df[output_colname] = np.where(df.loc[:,input_colname].str.contains('-', regex=False)==False, 
                               df.loc[:,input_colname].str[:-1]+"-"+df.loc[:,input_colname].str[-1:], 
                               df.loc[:,input_colname])
    
    df[output_colname] = np.where(df.loc[:,output_colname].str.contains('.', regex=False)==True, 
                            df.loc[:,output_colname].replace('.', ''), 
                            df.loc[:,output_colname])
    return df.head(3)
```

## Spatial data manipulation

Transformation of coordinates from UTM coordinates to longitude, latitude data based on: https://ocefpaf.github.io/python4oceanographers/blog/2013/12/16/utm/

```python
from pyproj import Proj

x = df_raw.loc[:,'COORD_E']
y = df_raw.loc[:,'COORD_N']

df_utm = pd.DataFrame(np.c_[x, y], columns=['Meters East', 'Meters South'], index=df_raw.index)

np.c_[x, y], df_utm.head()
```

```python
# Proj element to classify the projection of the data from the input
myProj = Proj(proj='utm', zone='18', south=True, ellps='WGS84', units='m')

lon, lat = myProj(df_utm.loc[:,'Meters East'].values, df_utm.loc[:,'Meters South'].values, inverse=True) 
# inverse=True refers to the conversion UTM-> lat,lon. inverse=False would mean lat,lon->UTM transformation

df_lat_lon = pd.DataFrame(np.c_[lat, lon], columns=['Lat', 'Lon'], index=df_utm.index)
df_lat_lon.head()
```

## Data selection and processing from cleaned data

```python
df = df_raw.merge(df_lat_lon, how='outer', left_on=df_raw.index, right_on=df_lat_lon.index)
df.head()
```

We create a function that selects the species of fruit to be analyse. If the fruit is acknowledged in one column then we assume it is not in the other columns **(it is either in one column or none)**.

We create a new column called 'superficie de viñas (has)' with the amount of acres (has) of vineyards per id.

```python
def species_selection(df ,species):
    df_especie1 = df[df.loc[:,'Frutales Mayores Especie 1'] == species]
    df_especie1.loc[:,'supercifie de {} (has)'.format(species.lower())] = df_especie1.loc[:,'Frutales Mayores Superfice especie 1 (has)']
    df_especie2 = df[df.loc[:,'Frutales Mayores Especie 2'] == species]
    df_especie2.loc[:,'supercifie de {} (has)'.format(species.lower())] = df_especie2.loc[:,'Frutales Mayores Superfice especie 2 (has)']
    df_especie3 = df[df.loc[:,'Frutales Mayores Especie 3'] == species]
    df_especie3.loc[:,'supercifie de {} (has)'.format(species.lower())] = df_especie3.loc[:,'Frutales Mayores Superfice especie 3 (has)']

    df_species = df_especie1.append([df_especie2, df_especie3])
    
    return df_species
```

```python
df_viñas = species_selection(df, 'VIÑAS')
```

## Data transformation from DataFrame to Network

```python
import networkx as nx
import matplotlib.pyplot as plt

G=nx.Graph()

# dictionaries for node attributes
coord_zip = zip(df_viñas.loc[:,'Lon'],df_viñas.loc[:,'Lat'])
node_dict_coord = dict(zip(df_viñas.index, coord_zip))
node_dict_comuna = dict(zip(df_viñas.index, df_viñas.loc[:,'COMUNA']))
node_dict_ha = dict(zip(df_viñas.index, df_viñas.loc[:,'supercifie de viñas (has)']))

# append dictionaries to add them as node attributes
from collections import defaultdict
dd = defaultdict(list)
d1 = node_dict_coord
d2 = node_dict_comuna
d3 = node_dict_ha
for d in (d1, d2, d3): # you can list as many input dicts as you want here
    for key, value in d.items():
        dd[key].append(value)
        
G.add_nodes_from([(node, {'location': attr[0], 'area': attr[1], 'supercifie de viñas (has)': attr[2]}) for (node, attr) in dd.items()])
```

```python
from itertools import count
area = set(nx.get_node_attributes(G,'area').values())
mapping = dict(zip(sorted(area),count()))
colors = [mapping[G.nodes[n]['area']] for n in G.nodes()]
```

Which is an inneficient and unreliable method. So the option is to **automatize** and drop values that are outsiders. This automatization requires **spatial boundaries** so we need to continue with the data from the region in study and use it to classify our data from within.

```python
# Plotting again along with the manual corrections for the erroneous data.

G.remove_nodes_from(lst_err_ids)

from itertools import count
area = set(nx.get_node_attributes(G,'area').values())
mapping = dict(zip(sorted(area),count()))
colors = [mapping[G.nodes[n]['area']] for n in G.nodes()]

node_size = [20*nx.get_node_attributes(G, 'supercifie de viñas (has)')[v] for v in G]

plt.figure(figsize=(10,7))


pos = nx.get_node_attributes(G, 'location')
nx.draw_networkx(G, pos, with_labels=False, node_size=node_size,
                node_color=colors, cmap=plt.cm.jet)

plt.axis('off')
plt.tight_layout();
```

As we can see, we could no longer take off manually the outsiders from the boundaries of the region as we took them off before (we only took off the ones that are far away from the mean), **AND** there are points in the plot that go wrong by having an assignment for one place of the map (by its attribute 'area') and in the map appears in another area due to their 'location' attribute. 

**This inconsistency compels us to find a way to select only the data that is within the map region AND its assignment matches both the map 'area' and the point 'area' where it belongs.**

```python
from shapely.geometry import Point, Polygon
```

```python
df_ñuble_geo = df_ñuble.loc[:,['Comuna', 'geometry']]
df_ñuble_geo.set_index('Comuna', inplace=True)
df_ñuble_geo.head()
```
