```python
def rut_generator():
    return str(np.random.randint(0, 1))+str(np.random.randint(1, 9))\
    +str(np.random.randint(0, 9))+str(np.random.randint(0, 9))+str(np.random.randint(0, 9))\
    +str(np.random.randint(0, 9))+str(np.random.randint(0, 9))+str(np.random.randint(0, 9))+'-'\
    +str(np.random.randint(1, 9))
rut = []
for i in range(1358):
    rut.append(rut_generator())
rut
```

```python
comunas = ['QUILLÓN', 'BULNES', 'PINTO', 'COELEMU', 'RANQUIL', 'TREHUACO',
       'EL CARMEN', 'SAN IGNACIO', 'COBQUECURA', 'NINHUE', 'QUIRIHUE',
       'SAN FABIAN', 'ÑIQUEN', 'SAN CARLOS', 'PEMUCO', 'YUNGAY',
       'SAN  NICOLAS', 'PORTEZUELO', 'Portezuelo', 'COIHUECO',
       'CHILLAN VIEJO', 'CHILLAN']
```

```python
def coord_generator():
    #return in COORD_E-COORD_N
    return (
        str(7)+str(np.random.randint(0, 9))\
    +str(np.random.randint(0, 9))+str(np.random.randint(0, 9))+str(np.random.randint(0, 9))\
    +str(np.random.randint(0, 9)),
            
        str(5)+str(9)\
    +str(np.random.randint(1, 9))+str(np.random.randint(0, 9))+str(np.random.randint(0, 9))\
    +str(np.random.randint(0, 9))+str(np.random.randint(0, 9)))
coord = []
for i in range(1358):
    coord.append(coord_generator())
coord = [(float(E), float(N)) for E, N in coord]
coord
```

```python
def value_generator(): #xx mm
    return (np.random.randint(100000, 9000000))
value = []
for i in range(1358):
    value.append(value_generator())
value
```

```python
def incentive_generator(): #xx mm
    return (np.random.randint(100000, 3000000))
incentive = []
for i in range(1358):
    incentive.append(incentive_generator())
```

```python
df_data1['COMUNA'] = np.random.choice(comunas, size=len(df_data1))
df_data1['RUT TITULAR'] = rut
df_data1['COORD_E'] = list(list(zip(*coord))[0])
df_data1['COORD_N'] = list(list(zip(*coord))[1])
```

```python
df_data1.to_excel('data_1.xlsx')
```

```python
df_data2 = pd.DataFrame()
df_data2['RUT'] = df_data1.loc[:,'RUT TITULAR']
df_data2['Total Proyecto'] = value
df_data2['Incentivo Solicitado'] = incentive
```
